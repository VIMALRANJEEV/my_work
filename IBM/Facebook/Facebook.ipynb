    Null hypothessis = The average likes of people who like a post on saturday is equal to people who like post on sunday.

    Alternate hypothesis = The average likes of people who like a post on saturday is not equal to people who like post on sunday.

    For the hypothesis testing we use paired t-test method. 

```python
a=fb[fb['Post Weekday']==6].like
b=fb[fb['Post Weekday']==7].like
a_mean=np.mean(a)
b_mean=np.mean(b)
print(a_mean,b_mean)
```

```python
from scipy.stats import skew,ttest_ind
t_statistic, p_value = ttest_ind(a , b)

print("t statistic = ",t_statistic," p value = ", p_value)
if p_value<0.05:
    print("The p value of",p_value,"is significant. Hence reject null hypothesis")
else:
    print("The p value of",p_value,"is not significant. Hence accept null hypothesis")
```
